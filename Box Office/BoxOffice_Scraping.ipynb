```python
from selenium.webdriver.common.by import By
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW")
```

```python
df=pd.DataFrame([],columns=['Rank','Title','Lifetime Gross','Year'])
Movies = driver.find_elements(By.XPATH,"//tr")
for movie in Movies:
    try:
        rank=movie.find_element(By.XPATH,".//td[@class='a-text-right mojo-header-column mojo-truncate mojo-field-type-rank']")
        print(rank.text)
        title=movie.find_element(By.XPATH,".//td[@class='a-text-left mojo-field-type-title']")
        print(title.text)
        lifetime_gross=movie.find_element(By.XPATH,".//td[@class='a-text-right mojo-field-type-money']")
        print(lifetime_gross.text)
        year=movie.find_element(By.XPATH,".//td[@class='a-text-left mojo-field-type-year']")
        print(year.text)
        df.loc[len(df)]=[rank.text,title.text,lifetime_gross.text,year.text]
    except:
        print('error')
```
