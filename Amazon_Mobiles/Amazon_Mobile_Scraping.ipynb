```python
from selenium.webdriver.common.by import By
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.amazon.eg/s?rh=n%3A21832883031&fs=true&language=en&ref=lp_21832883031_sar")
```

```python
mobiles = driver.find_elements(By.XPATH,"//span[@class='a-size-base-plus a-color-base a-text-normal']")
print("items=", len(mobiles))
for i in mobiles:
    print(i.text)
```

```python
price = driver.find_elements("xpath", '//span[@class="a-price-whole"]')

print("items=", len(price))

for i in price:
    print(i.text)
```

```python
rates = driver.find_elements("xpath", "//a[@class='a-popover-trigger a-declarative']")
print("items=", len(rates))
for i in rates:
    print(i)
```

```python
img = driver.find_elements("xpath", "//div[@class='a-section aok-relative s-image-square-aspect']")
print("items=", len(img))
for i in img:
    print(i)
```

```python
import urllib
df=pd.DataFrame([],columns=['a','b','c','d','e'])
products_Sections = driver.find_elements(By.XPATH,"//div[@class='a-section a-spacing-base']")
for section in products_Sections:
    try:
        title=section.find_element(By.XPATH,".//span[@class='a-size-base-plus a-color-base a-text-normal']")
        print(title.text)
        price=section.find_element(By.XPATH,".//span[@class='a-price-whole']")
        print(price.text)
        image_src=section.find_element(By.XPATH,".//img[@class='s-image']")
        url=image_src.get_attribute('src')
        print(url)
        rating=section.find_element(By.XPATH,".//div[@class='a-row a-size-small']/span").get_attribute('aria-label')
        print(rating.split()[0])
        total_rating=section.find_element(By.XPATH,".//span[@class='a-size-base s-underline-text']")
        print(total_rating.text)
        urllib.request.urlretrieve(url,"{}.jpg".format(title.text))
        df.loc[len(df)]=[title.text,price.text,url,total_rating.text,rating.split()[0]]
    except:
        print('error')
```
