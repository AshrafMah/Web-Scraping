```python
from selenium.webdriver.common.by import By
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.amazon.in/gp/bestsellers/books/")
```

```python
books = driver.find_elements(By.XPATH,'//div[@class="_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y"]')
print("items=", len(books))
for i in books:
    print(i.text)
```

```python
# There is a discrepancy between the number of results obtained from the XPath query directly (50 results) and the number of results obtained through Selenium (30 results).
price = driver.find_elements("xpath", '//span[@class="_cDEzb_p13n-sc-price_3mJ9Z"]')

print("items=", len(price))

for i in price:
    print(i.text)
```

```python
df=pd.DataFrame([],columns=['Title','Price'])
products_Sections = driver.find_elements(By.XPATH,"//div[@id='gridItemRoot']")
print(products_Sections)
for section in products_Sections:
    try:
        title=section.find_element(By.XPATH, './/div[@class="_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y"]')
        print(title.text)
        price=section.find_element(By.XPATH, './/span[@class="_cDEzb_p13n-sc-price_3mJ9Z"]')
        print(price.text)
        df.loc[len(df)]=[title.text,price.text]
    except:
        print('error')
```
